```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set(color_codes = True)
import urllib.request as ur
url = 'https://api.covid19india.org/states_daily.json'

ur.urlretrieve(url, "data.json")
covid_data = pd.read_json('data.json')

import json
with open ('data.json') as f:
  data = json.load(f)
data

data = data['states_daily']
cd = pd.json_normalize(data)
cd.drop('dateymd', axis =1, inplace = True)
```

```python
cd_kl=cd[['kl','date','status']]
# cd_kl.info()
print(cd_kl)
```

```python
cd_kl['kl']=pd.to_numeric(cd_kl['kl'])
cd_kl['date']=pd.to_datetime(cd_kl['date'])
```

```python
cd_kl=cd_kl.pivot_table(values='kl',columns='status', index='date')
cd_kl.head()
```

```python
sns.lineplot(x= 'date', y='Confirmed',data=cd_kl )
sns.lineplot(x= 'date', y='Recovered',data=cd_kl )
sns.lineplot(x= 'date', y='Deceased',data=cd_kl )
```

```python
cd_kl.plot.area()
```

```python
fig = plt.gcf()
fig.set_size_inches(12,6)
plt.stackplot(cd_kl.index, cd_kl.Confirmed,cd_kl.Recovered, cd_kl.Deceased, colors=['orange','green','red'], labels =('Confirmed','Recovered','Deceased'))
plt.legend()
```

```python
fig = plt.gcf()
fig.set_size_inches(20,12)
plt.stackplot(cd_kl.index, cd_kl.Confirmed,cd_kl.Deceased, cd_kl.Recovered, colors=['orange','green','red'], labels =('Confirmed','Deceased','Recovered'))
plt.legend()
```

```python
fig = plt.gcf()
fig.set_size_inches(12,8)
plt.stackplot(cd_kl.index, cd_kl.Confirmed/cd_kl.sum(axis=1), cd_kl.Recovered/cd_kl.sum(axis=1),cd_kl.Deceased/cd_kl.sum(axis=1), colors=['orange','green','red'], labels =('Confirmed','Deceased','Recovered'))
plt.legend()
```

```python
#state wise data using a function.
def plot_stack_area(state):
  cd_st = cd[[state,'date','status']]
  cd_st[state]=pd.to_numeric(cd_st[state])
  cd_st['date']=pd.to_datetime(cd_st['date'])
  cd_st = cd_st.pivot_table(values=state, columns='status', index='date')
  fig = plt.gcf()
  fig.set_size_inches(12,7)
  plt.stackplot(cd_st.index, cd_st.Confirmed,cd_st.Deceased, cd_st.Recovered, colors=['orange','green','red'], labels =('Confirmed','Deceased','Recovered'))
  plt.legend()
```

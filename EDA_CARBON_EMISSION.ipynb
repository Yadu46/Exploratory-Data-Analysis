```python
def outlier(data,column):
    Q3 = data[column].quantile(0.75)
    Q1 = data[column].quantile(0.25)
    IQR = Q3-Q1
    data = data[~((data[column]<(Q1-1.5*IQR))|(data[column]>(Q3+1.5*IQR)))]
    return data
```

```python
# The top countries which are contributing to the overal emissions of the global green house gases.
df_cem1=df_cem.sort_values(by='CO2_emission_1990_MT',ascending=False)
```

```python
df_cem1.index=np.arange(len(df_cem['Country']))
```

```python
df_cem1.head(11)
```

```python
# Visualization PLOTS
```

```python
# BAR Plot to check the contribution of first 10 countries in the global GHG emissions 
```

```python
# Top 10 countries in 1990 carbon Emission 
fig = plt.figure(figsize=(25,12))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
bar=sns.barplot(x=df_cem1.loc[1:11,'Country'], y=df_cem1.loc[1:11,'CO2_emission_1990_MT'],palette=sns.color_palette("flare", 10))
bar.set_xticklabels(bar.get_xticklabels(),rotation=60)
plt.xlabel('Top 10 Countries',fontsize=20,)
plt.ylabel('C02 Emissions in Metric Tonnes Y=1990',fontsize=20)
plt.annotate('8th position',xy=(7,800),xytext=(7.6,1500),arrowprops =dict(facecolor='Red', shrink=0.05),size=20)
```

```python
df_cem2=df_cem.sort_values(by='CO2_emission_2005_MT',ascending=False)
df_cem2.index=np.arange(len(df_cem['Country']))
df_cem2.head(11)
```

```python
# Top 10 countries in 2005 carbon Emission 
fig = plt.figure(figsize=(25,12))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
bar=sns.barplot(x=df_cem2.loc[1:11,'Country'], y=df_cem2.loc[1:11,'CO2_emission_2005_MT'],palette=sns.color_palette("rocket", 10))
bar.set_xticklabels(bar.get_xticklabels(),rotation=60)
plt.xlabel('Top 10 Countries',fontsize=20,)
plt.ylabel('C02 Emissions in Metric Tonnes Y=1990',fontsize=20)
plt.annotate('6th position',xy=(5.2,1300),xytext=(5.5,2000), arrowprops =dict(facecolor='Red', shrink=0.05),size=20)
```
